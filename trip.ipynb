```python
#scrapping data from Trip advisor
import requests             
from bs4 import BeautifulSoup 
import sys
import csv                  
import webbrowser
import io
import re
import pandas as pd
```

```python
url='https://www.tripadvisor.in/Restaurant_Review-g304551-d13388460-Reviews-Kitchen_With_A_Cause-New_Delhi_National_Capital_Territory_of_Delhi.html'
```

```python
# response= requests.get(url)
# data = response.text
# soup = BeautifulSoup(data,'html.parser')
# reviews = soup.find_all('div',{'class':'review-container'})
```

```python
task_no =0
review_data={}
c_page=0
```

```python
while True:
    
    response= requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'html.parser')
    reviews = soup.find_all('div',{'class':'review-container'})

    c_page =c_page + 1
    print(c_page,'\n')
    for review in reviews:
        ratingdates = review.find('span',{'class':'ratingDate'}).get('title')
        ratings = float(review.find('span',{'class':re.compile(r'ui_bubble_rating bubble_*')}).get("class")[1].split("_")[1])/10
        rating_title = review.find('span',{'class':'noQuotes'}).text
        description = review.find('p',{'class':'partial_entry'}).text
        site= 'Trip Advisor'
        task_no=task_no + 1
        review_data[task_no] = [ratings, ratingdates, description,rating_title,site]
        print('Ratings:', ratings, '\nDates:', ratingdates, '\nDescription:', description,  '\nTitle:',  rating_title, '\nSite:', site ,'\n---')
    

        
    url_tag = soup.find('a',{'class':'nav next ui_button primary'})
    try:
        if url_tag.get('href'):
            url='https://www.tripadvisor.in/' + url_tag.get('href')
            print(url)
    except AttributeError:
        if c_page == 20:
            break
         
 
    
```

```python
review_df = pd.DataFrame.from_dict(review_data, orient = 'index', columns = ['Ratings','Review Date', 'Description', 'Review Title', 'Site' ])
```
